```python
import re
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

import numpy as np
def clean_text(text):
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  
    cleaned_text = cleaned_text.strip()  
    cleaned_text = cleaned_text.lower()
    return cleaned_text

def tokenize_text(text):
    words = word_tokenize(text)
    return words


# a, an, and, are, as, at, be, but, by, for, if, in, into, is, it, no, not, of, on, or, such
def remove_stopwords(words):
    stop_words = set(stopwords.words('english'))
    filtered_words = [i for i in words if i not in stop_words]
    return filtered_words

# was -> be
# changing -> change ...
def lemmatize_words(words):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return lemmatized_words

def scrape_text_from_url(url):
    
    response = requests.get(url)
    response.raise_for_status()  
    if response.status_code != 200: return ""
    
    soup = BeautifulSoup(response.text, 'html.parser')

    text = soup.get_text()
    print("before")
    print(text)
    cleaned_text = clean_text(text)

    
    tokens = tokenize_text(cleaned_text)
    
    # Remove stop words
    filtered_tokens = remove_stopwords(tokens)

    # Lemmatize or stem the tokens
    ret = lemmatize_words(filtered_tokens)

    return ret

# Example usage:
url = "https://www.ibm.com/topics/artificial-intelligence"  # Replace with your URL
tokens = scrape_text_from_url(url)
#https://en.wikipedia.org/wiki/Artificial_intelligence
```
